```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
x
```

```python
y=2*x
y
```

# 1: Line plot

```python
plt.plot(x,y)
plt.show()
```

# -Adding label and title

```python
plt.plot(x,y)
plt.title("Line Plot")
plt.xlabel("x-label")
plt.ylabel("y-label")


plt.show()
```

# -Changing Line Aesthetics

```python
plt.plot(x,y,color='r',linestyle=':',linewidth=5)
```

# Adding two line in same plot

```python
x=np.arange(1,11)
y1=2*x
y2=3*x
```

plt.plot(x,y1,color='g',linestyle=':',linewidth='3')
plt.plot(x,y2,color='r',linestyle=':',linewidth='5')
plt.title("Line Plot")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.grid(True)
plt.show()

# Adding sub plots

```python
x=np.arange(1,11)
y1=2*x
y2=3*x

plt.subplot(2,1,1)
plt.plot(x,y1,color='g',linestyle=':',linewidth=2)

plt.subplot(2,1,2)
plt.plot(x,y2,color='r',linestyle=':',linewidth=2)
plt.title("Line Plot")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.grid(True)
plt.show()
```

# 2:Bar Plot

```python
student={"Mahesh":87,"Shivam":56,"Sam":27,"Raj":57}
```

```python
names=list(student.keys())
values=list(student.values())
```

```python
plt.bar(names,values)
plt.show()
```

# -Adding title and labels

```python
plt.bar(names,values,color='y')
plt.title("Distrubution of student marks")
plt.xlabel("Names of students")
plt.ylabel("Marks of students")
plt.grid(True)
plt.show()
```

# -Horizontal Bar Plot

```python
plt.barh(names,values,color='g')
plt.title("Distrubution of student marks")
plt.xlabel("Names of students")
plt.ylabel("Marks of students")
plt.grid(True)
plt.show()
```

# 3:Scatter Plot

```python
x=[10,20,30,40,50,60,70,80,90]
a=[9,4,8,6,2,9,3,7,1]

plt.scatter(x,a)
plt.show()
```

# -changing Mark Aesthetics

```python
x=[10,20,30,40,50,60,70,80,90]
a=[9,4,8,6,2,9,3,7,1]

plt.scatter(x,a,marker="*",c='g',s=100)
plt.show()
```

# -Adding Two Markers in same Plot

```python
x=[10,20,30,40,50,60,70,80,90]
a=[9,4,8,6,2,9,3,7,1]
b=[2,7,9,4,6,9,1,4,6]
plt.scatter(x,a,marker="*",c='r',s=100)
plt.scatter(x,b,marker=".",c='b',s=200)

plt.show()
```

# -Adding sub plots

```python
x=[10,20,30,40,50,60,70,80,90]
a=[9,4,8,6,2,9,3,7,1]
b=[2,7,9,4,6,9,1,4,6]

plt.subplot(1,2,1)
plt.scatter(x,a,marker="*",c='r',s=100)
plt.grid(True)

plt.subplot(1,2,2)
plt.scatter(x,b,marker=".",c='b',s=200)
plt.grid(True)
plt.show()
```
